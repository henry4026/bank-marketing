```python
import pandas as pd
import numpy as np

%load_ext autoreload
%autoreload 2
```

```python
data = pd.read_csv('bank-additional-full.csv',sep=';')
```

```python
new_data = data.copy()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.y.unique()
```

```python
## one hot encoding
```

```python
new_data.y[new_data.y == 'yes'] = 1 # convery label column to int
new_data.y[new_data.y == 'no'] = 0 # convery label column to int
```

```python
new_data = new_data.astype({'y':'int64'})
```

```python
new_data.info()
```

```python
new_data_df = pd.get_dummies(new_data, columns=['job','marital','education','day_of_week','default','housing','loan',
                                 'contact','month','poutcome'], drop_first=True)
```

```python
new_X = new_data_df.loc[:,  new_data_df.columns != 'y'] # select X
```

```python
new_y = new_data_df.y
new_X
```

```python
new_X.values
```

## train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(new_X.values,new_y.values,test_size=0.2,random_state=0)
```

##  resample

```python
!pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=12, sampling_strategy = 1)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
```

```python
np.unique(y_train_res, return_counts = True) # oversampling
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score
from sklearn.metrics import f1_score
```

```python
logistic_model = LogisticRegression(max_iter= 100,random_state=0)
```

```python
logistic_model.fit(X_train_res,y_train_res)
```

```python
logistic_model.score(X_train_res,y_train_res)
```

```python
pred = logistic_model.predict(X_test)
print(f"accuracy = {logistic_model.score(X_test, y_test)}")
print(f"r2 = {r2_score(y_test, pred)}")
print(f"f1 = {f1_score(y_test,pred)}")
```

```python
from sklearn.metrics import confusion_matrix
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, pred).ravel()
print(np.array([[tp, fn], [fp, tn]]))
```

```python
print("precision:{precision}, recall:{recall}".format(precision = tp/(tp+fp),recall = tp/(tp+fn)))
```

## Neural Network

```python
import tensorflow as tf
from tensorflow.keras import layers
keras = tf.keras
```

```python
nn_model = tf.keras.models.Sequential()
```

```python
nn_model.add(keras.layers.Dense(units=100, input_dim=53,activation='relu'))
nn_model.add(keras.layers.Dense(units=50, activation='relu'))
nn_model.add(keras.layers.Dense(units=20, activation='relu'))
```

```python
nn_model.add(keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
nn_model.compile(loss='binary_crossentropy',
              optimizer='adam',
                metrics=['accuracy'])
```

```python
nn_model.fit(X_train_res, y_train_res, epochs=10, batch_size=32)
```
